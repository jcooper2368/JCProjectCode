```python
#determine if missing income values are related to employment and age 

# Select the rows where income is missing
missing_income = medical_raw_data[medical_raw_data['Income'].isnull()]

# Group the missing income data by employment status
grouped = missing_income.groupby(['Employment', pd.cut(missing_income['Age'], bins=[0, 18, 30, 50, np.inf])])

# Count the number of missing income values for each employment status
missing_count = grouped.size()

# Calculate the percentage of missing income values for each employment status
percentage_missing = missing_count / medical_raw_data.groupby(['Employment', pd.cut(medical_raw_data['Age'], bins=[0, 18, 30, 50, np.inf])]).size() * 100

# Display the results
print(percentage_missing)
```

```python
#determine if missing no. of children values are related to martial status  

# Select the rows where no. of chilren is missing
missing_income = medical_raw_data[medical_raw_data['Children'].isnull()]

# Group the missing no. of children data by employment status
grouped = missing_income.groupby(['Marital', pd.cut(missing_income['Age'], bins=[18, 30, 50, np.inf])])

# Count the number of missing no. of children values for each marital status
missing_count = grouped.size()

# Calculate the percentage of missing income values for each employment status
percentage_missing = missing_count / medical_raw_data.groupby(['Marital', pd.cut(medical_raw_data['Age'], bins=[0, 18, 30, 50, np.inf])]).size() * 100

# Display the results
print(percentage_missing)
```

```python
# Select the rows where the Soft_drink column is null
missing_soft_drink = medical_raw_data[medical_raw_data['Soft_drink'].isnull()]

# Group the missing Soft_drink data by age
grouped = missing_soft_drink.groupby(pd.cut(missing_soft_drink['Age'], bins=[18, 30, 50, np.inf]))

# Count the number of missing Soft_drink values for each age group
missing_count = grouped.size()

# Calculate the percentage of missing Soft_drink values for each age group
percentage_missing = missing_count / medical_raw_data.groupby(pd.cut(medical_raw_data['Age'], bins=[0, 18, 30, 50, np.inf])).size() * 100

# Display the results
print(percentage_missing)
```

```python
from fancyimpute import KNN

# Define the columns to impute
columns_to_impute = ['Children', 'Income', 'Age', 'Initial_days', 'Population']

# Create an instance of the KNN imputer
knn_imputer = KNN()

# Get the column indices of the columns to impute
columns_to_impute_indices = [medical_raw_data.columns.get_loc(col) for col in columns_to_impute]

# Perform imputation on the selected columns
imputed_values = knn_imputer.fit_transform(medical_raw_data.iloc[:, columns_to_impute_indices])

# Assign the imputed values back to the original dataset
medical_raw_data.iloc[:, columns_to_impute_indices] = imputed_values

# Verify if the imputations are in the original dataset
print(medical_raw_data.head())
```

```python
#pca 

medical_raw_data_pca = medical_raw_data [[ 'Population','Children', 'Age', 'Income','VitD_supp', 'Initial_days', 'TotalCharge', 'Additional_charges', 'Item1', 'Item2', 'Item3', 'Item4', 'Item5', 'Item6', 'Item7', 'Item8']]  
medical_raw_data_pca.head()
```

```python
from sklearn.decomposition import PCA

#normalize the data 
pca_normalized=(medical_raw_data_pca-medical_raw_data_pca.mean())/medical_raw_data_pca.std()

pca = PCA(n_components=medical_raw_data_pca.shape[1])

#convert the variables to components 
pca.fit(pca_normalized)
medical_pca = pd.DataFrame(pca.transform(pca_normalized),     
     columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13','PC14','PC15','PC16'])
```

```python
medical_pca.head()
```

```python
#make scree plot 

plt.plot(pca.explained_variance_ratio_)
plt.xlabel('number of components')
plt.ylabel('explained variance')
plt.show()
```

```python
loadings = pd.DataFrame(pca.components_.T,
     columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13','PC14','PC15','PC16'],
     index=medical_raw_data_pca.columns)
loadings
```
