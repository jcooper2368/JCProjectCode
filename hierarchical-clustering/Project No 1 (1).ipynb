```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.metrics import silhouette_score

# Step 1: Compute the linkage matrix using customer_data_sampled
Z = linkage(customer_data_sampled, method='ward')  

# Step 2: Calculate the total within-cluster variance for different numbers of clusters
max_clusters = 10  # Set a maximum number of clusters you want to consider
total_within_cluster_variance = []

for k in range(2, max_clusters + 1):
    cluster_labels = fcluster(Z, k, criterion='maxclust')
    
    # Calculate within-cluster variance (inertia or sum of squared distances)
    variance = 0
    for i in range(1, k + 1):
        cluster_points = customer_data_sampled[cluster_labels == i]
        cluster_center = cluster_points.mean(axis=0)
        variance += np.sum((cluster_points - cluster_center) ** 2)
    
    total_within_cluster_variance.append(variance)

# Step 3: Plot number of clusters vs. within-cluster variance
plt.plot(range(2, max_clusters + 1), total_within_cluster_variance, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Total within-cluster variance')
plt.title('Elbow Method for Optimal Clusters')
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Define the number of clusters to evaluate
cluster_numbers = [3 ,4, 5, 6]
silhouette_scores = []

for n_clusters in cluster_numbers:
    # Fit the KMeans model
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(X)

    # Calculate silhouette score
    score = silhouette_score(X, cluster_labels)
    silhouette_scores.append(score)
    print(f'Silhouette score for {n_clusters} clusters: {score:.4f}')

# Plot silhouette scores for comparison
plt.figure(figsize=(8, 4))
plt.bar(cluster_numbers, silhouette_scores, color='skyblue')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different Cluster Numbers')
plt.xticks(cluster_numbers)
plt.ylim(0, 1)
plt.show()
```

```python
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import classification_report, make_scorer, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Create a customer-level dataset where each row represents a unique customer
customer_data = customer_data_sampled.groupby('Order_ID').agg(
    last_purchase_month=('purchase_month', 'max'),  # Most recent purchase month
    purchase_frequency=('Order_ID', 'count')  # Total purchase count
).reset_index()

# Define churn threshold at the customer level
current_month = customer_data_sampled['purchase_month'].max()
churn_threshold = 6  # Define your threshold in months
customer_data['churn'] = customer_data['last_purchase_month'].apply(
    lambda x: 1 if (current_month - x) > churn_threshold else 0
)

# Define features and target for model
X = customer_data[['purchase_frequency']]
y = customer_data['churn']

# Define the model
model = RandomForestClassifier(class_weight='balanced', random_state=42)

# Define ROC AUC score as a scorer for cross-validation
roc_auc_scorer = make_scorer(roc_auc_score)

# Perform cross-validation (e.g., 5-fold cross-validation)
cv_results = cross_validate(model, X, y, cv=5, scoring=roc_auc_scorer, return_train_score=False)

# Output cross-validation results
print("Cross-validation ROC AUC Scores: ", cv_results['test_score'])
print(f"Mean ROC AUC Score: {np.mean(cv_results['test_score'])}")
print(f"Standard Deviation of ROC AUC Scores: {np.std(cv_results['test_score'])}")

# Optionally, you can print the detailed classification report for the final model
model.fit(X, y)  # Train the final model on the entire dataset
y_pred = model.predict(X)
print(classification_report(y, y_pred))

# Check distribution of churn labels
print("Churn label distribution:")
print(customer_data['churn'].value_counts())
```

```python
#Predicting Customer Lifetime Value 
#CLV= avg order value * avg purchase frequency * avg customer lifespan

# Predicting Customer Lifetime Value
# CLV = avg order value * avg purchase frequency * avg customer lifespan

# Sample aggregated CLV calculation by cluster
clv_by_cluster = customer_data_sampled.groupby('cluster').agg(
    total_revenue=('Price_in_USD', 'sum'),  # Total revenue from purchases in the cluster
    purchase_frequency=('Order_ID', 'count'),  # Total number of orders in the cluster
    average_quantity=('Quantity_of_SKU_in_the_order', 'mean')  # Average quantity per order
).reset_index()

# Calculate average order value per cluster
clv_by_cluster['average_order_value'] = clv_by_cluster['total_revenue'] / clv_by_cluster['purchase_frequency']

# Assuming a churn rate (0.31 means the average customer lifespan is ~3.2 years)
churn_rate = 0.31
customer_lifespan = 1 / churn_rate  # Customer lifespan in years, inverse of churn rate

# Calculate CLV (Lifetime Value) for each cluster
clv_by_cluster['clv'] = clv_by_cluster['average_order_value'] * clv_by_cluster['purchase_frequency'] * customer_lifespan

# Convert CLV to millions and round to 2 decimal places
clv_by_cluster['clv'] = clv_by_cluster['clv'] / 1_000_000  # Convert to millions
clv_by_cluster['clv'] = clv_by_cluster['clv'].round(2)  # Round to 2 decimal places

# Optionally, format as string to add "M" for millions
clv_by_cluster['clv'] = clv_by_cluster['clv'].apply(lambda x: f"{x} M")

# View the result
print(clv_by_cluster[['cluster', 'clv']])

```

```python
# Merge the CLV data with the customer data based on the 'cluster' column
customer_data_sampled = customer_data_sampled.merge(
    clv_by_cluster[['cluster', 'clv']],  # Selecting only the 'cluster' and 'clv' columns
    on='cluster',  # Merge on the 'cluster' column
    how='left'  # Perform a left join to keep all rows in the customer_data_sampled
)

# View the updated customer_data_sampled DataFrame
print(customer_data_sampled.head())
```
